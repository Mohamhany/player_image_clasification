```python
image_urls = set() 
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # Find all images
    images = driver.find_elements(By.TAG_NAME, 'img')
    new_images = [image.get_attribute('src') for image in images]
    
    # Add new images to the set
    image_urls.update(new_images)

    # Scroll down to the bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait for new content to load
    time.sleep(5)  # Adjust delay as needed

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break  # Break the loop if no new content is loaded

    last_height = new_height

# Print all unique image URLs
print("Images URLs:", list(image_urls))
print("Number of scraped images:", len(image_urls))
```
